```python
def load_data(file):
    with open (file, "r", encoding="utf-8") as f:
        data = json.load(f) 
    return (data)

def write_data(file, data):
    with open (file, "w", encoding="utf-8") as f:
        json.dump(data, f, indent=4)
```

```python
#NLTK - Natural Language Tool-Kit for English 
import nltk 
nltk.download('stopwords')
#stopwords = stopwords.words("english")
stopwords = nltk.corpus.stopwords.words('english')
```

```python
print (stopwords)
```

```python
data = load_data("tf_41178_bodyOutput.json")
for text in data:
    print(text)
```

```python
#en_core_web_sm - includes syntax and vocabulary
import en_core_web_sm 
nlp = en_core_web_sm.load()

#Helps convert a document/sentances into tokens (seperate strings)
def sent_to_words(sentences):
    for sentence in sentences:
        yield(gensim.utils.simple_preprocess(str(sentence), deacc=True))  # deacc=True removes punctuations from the words

#Helps removes stopwords such as "is", "and", "if"... from all the processed words
def remove_stopwords(texts):
    return [[word for word in simple_preprocess(str(doc)) 
             if word not in stopwords] for doc in texts]

#Storing the tozenized words from sent_to_words into a list
data_words = list(sent_to_words(data))

#Storing the words without the stop words associated to them
data_words_nostops = remove_stopwords(data_words)

#Using the imported Spacy language model to apply linguistic features and rules
nlp = spacy.load('en_core_web_sm', disable=['parser', 'ner'])

#Do lemmatization (reducing the words ot its base form) keeping only noun, adj, vb, adv
data_lemmatized = []
for sent in data_words_nostops:
    doc = nlp(" ".join(sent)) 
    data_lemmatized.append([token.lemma_ for token in doc])

print(data_lemmatized)
```

```python
#Creates a Dictionary to map the token words to their respective IDs
id2word = corpora.Dictionary(data_lemmatized)
#Creates a list of vectors associated to the number of words
texts = data_lemmatized
corpus = [id2word.doc2bow(text) for text in texts]
# Viewing 5 bags of words shown as key values pairs of WordID and Frequency - can change this number as it is currently an arbitrary number
print(corpus[:5])
```

```python
# Build LDA model
lda_model = gensim.models.LdaMulticore(corpus=corpus,
                                       id2word=id2word,
                                       num_topics=4,  # using 4 to represent: ML review, Conventional review, Management, Other - (for later)
                                       random_state=100,
                                       chunksize=100,
                                       passes=10,
                                       per_word_topics=True)
print(lda_model.print_topics())
```

```python
#Optional Code: The Computation of the Coherance Score.
#A measure used to evaluate the quality of the topics produced by the topic model.
coherence_model_lda = CoherenceModel(model=lda_model, texts=data_lemmatized, dictionary=id2word, coherence='c_v')
coherence_lda = coherence_model_lda.get_coherence()
#The printed score is currently considered good.
print('\nCoherence Score: ', coherence_lda) 
```

```python
# Visualize the topics
pyLDAvis.enable_notebook()
vis = pyLDAvis.gensim_models.prepare(lda_model, corpus, id2word)
vis
```
